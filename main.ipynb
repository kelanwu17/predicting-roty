```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df_result = df
df = df.drop(columns='Player')
```

```python
X_train = df.loc[(df['Year'] < 2022)]
y_train = X_train['Share']

X_train = X_train.drop(columns='Share')
X_train

```

```python
y_train
```

```python
X_test = df.loc[(df['Year'] >= 2022)]
y_test = X_test['Share']

X_test = X_test.drop(columns=['Share'])
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

```

```python
X_test = df_result.loc[(df['Year'] >= 2022)]
X_test = X_test.drop(columns=['Share'])

X_test["Share Predicted"] = y_pred
X_test["Actual Share"] = y_test
X_test = X_test.sort_values(by="Share Predicted", ascending=False)
pd.set_option('display.max_rows', None)
X_test
```
