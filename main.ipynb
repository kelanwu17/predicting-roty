```python
import pandas as pd
```

```python
df = pd.read_csv('total_rookies.csv')
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df.info()
```

```python
print(df.isnull().values.any())
print(df.isnull().sum())
```

```python
for i in range(0,len(df)):
    if pd.isnull(df.loc[i,"FG%"]):
        df["FG%"][i] = df["FG"][i] / df["FGA"][i]
    if pd.isnull(df.loc[i,"3P%"]):
        df["3P%"][i] = df["3P"][i] / df["3PA"][i]
    if pd.isnull(df.loc[i,"FT%"]):
        df["FT%"][i] = df["FT"][i] / df["FTA"][i]

print(df.isnull().values.any())
print(df.isnull().sum())
```
