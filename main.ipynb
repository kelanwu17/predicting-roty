```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df_result = df
df = df.drop(columns=['Player','Pts Won','Pts Max','Tm','Team'])
```

```python
df.corr()['Share']
```

```python
year = 2023

X_train = df.loc[(df['Year'] < year)]
y_train = X_train['Share']
X_train = X_train.drop(columns='Share')

X_test = df.loc[(df['Year'] == year)]
y_test = X_test['Share']
X_test = X_test.drop(columns=['Share'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(r2)
print(mae)
print(mse)

X_test = df_result.loc[(df['Year'] == year)]
X_test = X_test.drop(columns=['Share'])

X_test["Share Predicted"] = y_pred
X_test["Actual Share"] = y_test
X_test = X_test.sort_values(by="Share Predicted", ascending=False)

pd.set_option("display.max_columns", None)
```
