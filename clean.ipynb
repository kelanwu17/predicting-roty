```python
import pandas as pd
```

```python
rotys = pd.read_csv('rotys.csv')
rotys.head()
```

```python
rotys = rotys[["Player","Year",'Pts Won','Pts Max','Share']]
rotys
```

```python
players = pd.read_csv('players.csv')
players.head()
```

```python
players["Player"] = players["Player"].str.replace("*","", regex=False)

players.head()
```

```python
players.drop(columns=['Unnamed: 0','Rk','Debut'])
```

```python
combined = players.merge(rotys, how="outer", on=["Player", "Year"])
combined
```

```python
combined[["Pts Won","Pts Max","Share"]] = combined[["Pts Won","Pts Max","Share"]].fillna(0)
combined
```

```python
combined = combined.drop(columns=["Unnamed: 0","Rk"])
```

```python
pd.set_option('display.max_columns', None)
combined.head()
```

```python
combined = combined.rename(columns={"MP.1": "MPG", "PTS.1":"PPG", "TRB.1":"RPG","AST.1":"APG"})
combined.head()
```

```python
import re
for i in range(len(combined['Debut'])):
    combined['Debut'][i] = re.sub(r'\d+','',combined['Debut'][i])
    combined['Debut'][i] = combined['Debut'][i].replace(',','')
    combined['Debut'][i] = combined['Debut'][i].replace('\'','')
    combined['Debut'][i] = combined['Debut'][i].replace('@','')
    combined['Debut'][i] = combined['Debut'][i].replace('vs.','')
    combined['Debut'][i] = combined['Debut'][i][6:-5]
    print(combined['Debut'][i])
```

```python
combined = combined.rename(columns={'Debut':'Tm'})
combined.head()
```

```python
teams = pd.read_csv('teams.csv')
teams
```

```python
teams["Team"] = teams["Team"].str.replace("*","",regex=False)
teams = teams.drop(columns="Unnamed: 0")

teams
```

```python
teams["Team"].unique()
```

```python
nicknames = {}
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
       abbrev,name = line.replace("\n","").split(",")
       nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
combined.head()
```

```python
stats = combined.merge(teams, how="outer", on=["Team", "Year"])

stats = stats.dropna(subset=["Player"])
stats
```

```python
combined
```

```python
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
stats["GB"].unique()
```

```python

stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats.to_csv("total_rookies_stats.csv")
```

```python
highest_scoring = combined[combined['G'] > 70].sort_values("PPG", ascending = False).head(10)

sns.barplot(data= highest_scoring, x="Player", y="PPG")
plt.xticks(rotation=45, fontsize=12)
plt.show()
```
