```python
import pandas as pd
```

```python
rotys = pd.read_csv('rotys.csv')
rotys.head()
```

```python
rotys = rotys[["Player","Year",'Pts Won','Pts Max','Share']]
rotys
```

```python
players = pd.read_csv('players.csv')
players.head()
```

```python
players["Player"] = players["Player"].str.replace("*","", regex=False)

players.head()
```

```python
players.drop(columns=['Unnamed: 0','Rk','Debut'])
```

```python
combined = players.merge(rotys, how="outer", on=["Player", "Year"])
combined.head()
```

```python
combined[combined["Pts Won"] > 0]
```

```python
combined[["Pts Won","Pts Max","Share"]] = combined[["Pts Won","Pts Max","Share"]].fillna(0)
combined
```

```python
combined = combined.drop(columns=["Unnamed: 0","Rk","Debut"])
```

```python
pd.set_option('display.max_columns', None)
combined.head()
```

```python
combined = combined.rename(columns={"MP.1": "MPG", "PTS.1":"PPG", "TRB.1":"RPG","AST.1":"APG"})
combined.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
highest_scoring = combined[combined['G'] > 70].sort_values("PPG", ascending = False).head(10)

sns.barplot(data= highest_scoring, x="Player", y="PPG")
plt.xticks(rotation=45, fontsize=12)
plt.show()
```

```python
highest_assist = combined[combined['G'] > 70].sort_values("APG", ascending = False).head(10)

sns.barplot(data= highest_assist, x="Player", y="APG")
plt.xticks(rotation=45, fontsize=12)
plt.show()
```

```python
highest_rebound = combined[combined['G'] > 70].sort_values("RPG", ascending = False).head(10)

sns.barplot(data= highest_rebound, x="Player", y="RPG")
plt.xticks(rotation=45, fontsize=12)
plt.show()
```

```python
combined.groupby("Year").apply(lambda x: x.sort_values("RPG", ascending = False).head(1))
```

```python
stat_corr = combined.corr()["Share"]
stat_corr
```
