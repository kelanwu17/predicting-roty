```python
dfs = []
for year in years:
    with open("roty/{}.html".format(str(year)), encoding="utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="over_header").decompose()
    roty_table = soup.find_all(id="roy")
    roty = pd.read_html(str(roty_table), header= 1)[0]
    roty["Year"] = year
    dfs.append(roty)
    
print(len(dfs))
```

```python
rotys = pd.concat(dfs)
rotys.head()
```

```python
rotys.to_csv("rotys.csv")
```

```python
df = []

for year in years:
    with open("players/{}.html".format(year),encoding="utf-8") as f:
        page = f.read()
        
    soup = BeautifulSoup(page, "html.parser")
    element = soup.find_all('tr', class_="over_header")
    for i in element:
        i.decompose()
    element = soup.find_all('tr', class_="thead")
    for i in element:
        i.decompose()
    player_table = soup.find_all(id="rookies")
    player = pd.read_html(str(player_table))[0]
    player["Year"] = year
    df.append(player)

```

```python
players = pd.concat(df)
players
```

```python
players.to_csv("players.csv")
```
